```python
# Week 3: Statistics & Probability
# AnalystLab Africa Internship
# Thembinkosi Spogter Phama

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline

# Load cleaned datasets
base_path = r'C:\Users\THEMBINKOSI\Desktop\AnalystLab_Internship\datasets'
titanic = pd.read_csv(f'{base_path}\\titanic_cleaned.csv')
housing = pd.read_csv(f'{base_path}\\housing_cleaned.csv')

print("✅ Week 3 ready!")
print(f"Titanic: {titanic.shape}")
print(f"Housing: {housing.shape}")
```

```python
print("=" * 60)
print("TITANIC - DESCRIPTIVE STATISTICS")
print("=" * 60)

# Numerical summary
print("\n📊 Numerical Features:")
display(titanic.describe())

# Categorical summary
print("\n📋 Categorical Features:")
display(titanic.describe(include='object'))

# Additional stats
print("\n📐 Additional Measures:")
for col in ['Age', 'Fare']:
    print(f"\n{col}:")
    print(f"  Mean:   {titanic[col].mean():.2f}")
    print(f"  Median: {titanic[col].median():.2f}")
    print(f"  Mode:   {titanic[col].mode()[0]:.2f}")
    print(f"  Std:    {titanic[col].std():.2f}")
    print(f"  Var:    {titanic[col].var():.2f}")
    print(f"  Skew:   {titanic[col].skew():.2f}")
    print(f"  Kurt:   {titanic[col].kurtosis():.2f}")
    print(f"  Range:  {titanic[col].max() - titanic[col].min():.2f}")
    print(f"  IQR:    {titanic[col].quantile(0.75) - titanic[col].quantile(0.25):.2f}")
```

```python
print("=" * 60)
print("HOUSING - DESCRIPTIVE STATISTICS")
print("=" * 60)

# Numerical summary
print("\n📊 Numerical Features:")
display(housing.describe())

# Categorical summary
print("\n📋 Categorical Features:")
display(housing.describe(include='object'))

# Skewness & Kurtosis for key features
print("\n📐 Distribution Shape:")
for col in ['price', 'area', 'bedrooms', 'bathrooms']:
    print(f"\n{col}:")
    print(f"  Skewness: {housing[col].skew():.2f}")
    print(f"  Kurtosis: {housing[col].kurtosis():.2f}")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Titanic - Distribution Analysis', fontsize=16, fontweight='bold')

# Age Distribution with Normal Curve
sns.histplot(titanic['Age'], bins=30, kde=True, ax=axes[0,0], color='#4ECDC4')
axes[0,0].axvline(titanic['Age'].mean(), color='red', linestyle='--', label=f'Mean: {titanic["Age"].mean():.1f}')
axes[0,0].axvline(titanic['Age'].median(), color='green', linestyle='--', label=f'Median: {titanic["Age"].median():.1f}')
axes[0,0].set_title('Age Distribution (Approx. Normal)')
axes[0,0].legend()

# Fare Distribution
sns.histplot(titanic['Fare'], bins=40, kde=True, ax=axes[0,1], color='#FF6B6B')
axes[0,1].axvline(titanic['Fare'].mean(), color='red', linestyle='--', label=f'Mean: {titanic["Fare"].mean():.1f}')
axes[0,1].axvline(titanic['Fare'].median(), color='green', linestyle='--', label=f'Median: {titanic["Fare"].median():.1f}')
axes[0,1].set_title('Fare Distribution (Right-Skewed)')
axes[0,1].legend()

# Q-Q Plot for Age
stats.probplot(titanic['Age'].dropna(), dist="norm", plot=axes[0,2])
axes[0,2].set_title('Q-Q Plot: Age vs Normal')

# Boxplot Age by Survival
sns.boxplot(data=titanic, x='Survived', y='Age', palette=['#FF6B6B', '#4ECDC4'], ax=axes[1,0])
axes[1,0].set_title('Age Distribution by Survival')
axes[1,0].set_xticklabels(['Did Not Survive', 'Survived'])

# Boxplot Fare by Class
sns.boxplot(data=titanic, x='Pclass', y='Fare', palette='Set2', ax=axes[1,1])
axes[1,1].set_title('Fare Distribution by Class')
axes[1,1].set_yscale('log')

# Survival Probability
surv_prob = titanic['Survived'].value_counts(normalize=True)
axes[1,2].bar(['Did Not Survive', 'Survived'], surv_prob.values, color=['#FF6B6B', '#4ECDC4'])
axes[1,2].set_title('Survival Probability Distribution')
for i, v in enumerate(surv_prob.values):
    axes[1,2].text(i, v+0.01, f'{v:.1%}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
print("=" * 70)
print("HYPOTHESIS TESTING")
print("=" * 70)

# ═══════════════════════════════════════════════
# TEST 1: Is there a significant difference in Age between Survivors and Non-Survivors?
# ═══════════════════════════════════════════════
print("\n" + "="*50)
print("TEST 1: Age Difference — Survivors vs Non-Survivors")
print("="*50)

survived_age = titanic[titanic['Survived']==1]['Age']
died_age = titanic[titanic['Survived']==0]['Age']

t_stat, p_value = stats.ttest_ind(survived_age, died_age)

print(f"Mean Age - Survived: {survived_age.mean():.2f}")
print(f"Mean Age - Died:     {died_age.mean():.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value:     {p_value:.4f}")
print(f"Result:      {'✅ SIGNIFICANT difference' if p_value < 0.05 else '❌ NOT significant'}")

# ═══════════════════════════════════════════════
# TEST 2: Is there a significant difference in Fare between Classes?
# ═══════════════════════════════════════════════
print("\n" + "="*50)
print("TEST 2: Fare Difference — 1st Class vs 3rd Class")
print("="*50)

first_fare = titanic[titanic['Pclass']==1]['Fare']
third_fare = titanic[titanic['Pclass']==3]['Fare']

t_stat, p_value = stats.ttest_ind(first_fare, third_fare)

print(f"Mean Fare - 1st Class: {first_fare.mean():.2f}")
print(f"Mean Fare - 3rd Class: {third_fare.mean():.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value:     {p_value:.4f}")
print(f"Result:      {'✅ SIGNIFICANT difference' if p_value < 0.05 else '❌ NOT significant'}")

# ═══════════════════════════════════════════════
# TEST 3: Chi-Square Test — Is Survival dependent on Gender?
# ═══════════════════════════════════════════════
print("\n" + "="*50)
print("TEST 3: Chi-Square — Survival vs Gender")
print("="*50)

contingency = pd.crosstab(titanic['Sex'], titanic['Survived'])
chi2, p_value, dof, expected = stats.chi2_contingency(contingency)

print("Contingency Table:")
print(contingency)
print(f"\nChi-Square: {chi2:.4f}")
print(f"P-value:    {p_value:.8f}")
print(f"Result:     {'✅ SIGNIFICANT dependency' if p_value < 0.05 else '❌ NOT significant'}")

# ═══════════════════════════════════════════════
# TEST 4: Housing — Is there a difference in Price by Air Conditioning?
# ═══════════════════════════════════════════════
print("\n" + "="*50)
print("TEST 4: Price Difference — AC vs No AC (Housing)")
print("="*50)

ac_yes = housing[housing['airconditioning']=='yes']['price']
ac_no = housing[housing['airconditioning']=='no']['price']

t_stat, p_value = stats.ttest_ind(ac_yes, ac_no)

print(f"Mean Price - AC:     {ac_yes.mean():,.0f}")
print(f"Mean Price - No AC:  {ac_no.mean():,.0f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value:     {p_value:.6f}")
print(f"Result:      {'✅ SIGNIFICANT difference' if p_value < 0.05 else '❌ NOT significant'}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Hypothesis Testing Visualizations', fontsize=16, fontweight='bold')

# Test 1: Age by Survival
sns.boxplot(data=titanic, x='Survived', y='Age', palette=['#FF6B6B', '#4ECDC4'], ax=axes[0,0])
axes[0,0].set_title('Age: Survivors vs Non-Survivors\n(p-value: compare to 0.05)', fontsize=11)
axes[0,0].set_xticklabels(['Died', 'Survived'])

# Test 2: Fare by Class
sns.boxplot(data=titanic, x='Pclass', y='Fare', palette='Set2', ax=axes[0,1])
axes[0,1].set_title('Fare by Passenger Class\n(1st vs 3rd — significant?)', fontsize=11)
axes[0,1].set_yscale('log')

# Test 3: Survival by Gender
contingency_plot = pd.crosstab(titanic['Sex'], titanic['Survived'], normalize='index')*100
contingency_plot.plot(kind='bar', ax=axes[1,0], color=['#FF6B6B', '#4ECDC4'], edgecolor='black')
axes[1,0].set_title('Survival Rate by Gender (%)\n(Chi-Square Test)', fontsize=11)
axes[1,0].set_xticklabels(['Female', 'Male'], rotation=0)
axes[1,0].legend(['Died', 'Survived'])
axes[1,0].set_ylabel('Percentage')

# Test 4: Housing AC
sns.boxplot(data=housing, x='airconditioning', y='price', palette=['#FF6B6B', '#4ECDC4'], ax=axes[1,1])
axes[1,1].set_title('House Price: AC vs No AC\n(T-Test)', fontsize=11)
axes[1,1].set_ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
print("=" * 70)
print("CORRELATION VS CAUSATION ANALYSIS")
print("=" * 70)

# Correlation matrix with interpretation
print("\n📊 TITANIC CORRELATIONS WITH SURVIVAL:")
titanic_corr = titanic[['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']].corr()['Survived'].sort_values(ascending=False)
print(titanic_corr)

print("\n📊 HOUSING CORRELATIONS WITH PRICE:")
housing_corr = housing[['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']].corr()['price'].sort_values(ascending=False)
print(housing_corr)

# ═══════════════════════════════════════════════
# CORRELATION vs CAUSATION EXPLANATION
# ═══════════════════════════════════════════════
print("\n" + "="*70)
print("CORRELATION ≠ CAUSATION")
print("="*70)

print("""
🔍 EXAMPLES FROM OUR DATA:

1. TITANIC: Fare is correlated with Survival (r=0.26)
   - CORRELATION: Higher fare → Higher survival
   - CAUSATION? NO — Fare is a proxy for Class.
   - True cause: 1st class passengers were prioritized for lifeboats.
   - Fare doesn't CAUSE survival; class status does.

2. HOUSING: Bedrooms correlated with Price (r=0.37)
   - CORRELATION: More bedrooms → Higher price
   - CAUSATION? Partially — but Area (r=0.54) is the real driver.
   - A small house with many tiny bedrooms won't be expensive.

3. TITANIC: Age vs Survival (r=-0.06, p=0.0528)
   - Very weak correlation, NOT statistically significant.
   - Children survived more, but age alone doesn't predict survival.
   - The "women and children first" policy is the real cause.

⭐ RULE: Just because two things move together doesn't mean one causes the other.
""")
```
